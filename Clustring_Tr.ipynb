```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(datafile)
```

```python
# look at "all" the data
df.describe()
```

```python
# feature engineering
df.drop(['pol_number', 'pol_eff_dt', 'gender', 'agecat', 'area', 'veh_age', 'veh_body', 'claim_office', 'annual_premium'], axis=1, inplace=True)
# map date_of_birth to age (years) then drop
from datetime import date
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
df['age']  = df['date_of_birth'].map(lambda x: (date(2020, 12, 31) - x.date()).days / 365)
df.drop(['date_of_birth'], axis=1, inplace=True)
df
```

```python
# remove outliers
outlier = df['veh_value'].quantile(0.99)
df = df[df['veh_value'] < outlier]
```

```python
# remove null values
print(df.isna().any(axis=1).sum())
df = df[~df.isna().any(axis=1)]
df.info()
```

```python
df
```

```python
# risky behavior translates into higher numclaims and claimcst0
# what are they for the entire population?
print(df.numclaims.mean())
print(df.claimcst0.mean())
df[['numclaims', 'claimcst0']].corr()
```

```python
# cluster predictive variables into 3 clusters
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.cluster import KMeans
from sklearn import metrics

X = df.copy()

model = KMeans(n_clusters=3, random_state=0).fit(Normalizer().fit_transform(X.drop(['numclaims', 'claimcst0'], axis=1)))

X['labels'] = model.labels_
```

```python
# figure out which class is low|average|high risk
print(X['labels'].value_counts())
print(X.groupby(['labels']).numclaims.mean())
print(X.groupby(['labels']).claimcst0.mean())
```
